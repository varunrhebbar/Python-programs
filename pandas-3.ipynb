```python
#pivot & pivot table

import pandas as pd
df=pd.read_csv("weather_by_cities.csv")
df
```

```python
df.pivot(index="day",columns="city")
```

```python
df.pivot(index="day",columns="city",values="temperature")
```

```python
df=pd.read_csv("iris.csv")
df
```

```python
df.pivot_table(index="variety",columns="sepal.length")
```

```python
df.pivot_table(index="variety",columns="sepal.length",aggfunc="mean")
```

```python
df.pivot_table(index="sepal.length",columns="variety",aggfunc="count",margins="True")
```
