```python
import pandas as pd
df=pd.read_csv("diabetes.csv")
#print(df)

groupby_dict={'Pregnancies  ':'col_1',
              'Glucose  ':'col_1',
              'BloodPressure   ':'col_1',
              'SkinThickness ':'col_1',
              'Insulin   ':'col_1',
              'BMI  ':'col_2',
              'DiabetesPedigreeFunction  ':'col_2',
              'Age  ':'col_2',
              'Outcome  ':'col_2'
            }
df= df.set_index('Pregnancies')
df= df.groupby(groupby_dict, axis=1).sum()
print(df)
```
